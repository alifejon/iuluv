```python
'''
drum_data :
    1. drum_data: (n.offset, p.midi, 1)  dict()  
    2. 

'''

drum_data = dict()

for n in o.parts[0].getElementsByClass(music21.chord.Chord):
    print('----')
    print("chord {} : {}".format(n, n.offset))
    print("notes {}".format(n.pitches))
    if n.offset not in drum_data:
        drum_data[n.offset] = dict()
    drum_data[n.offset]['offset'] = n.offset
    for p in n.pitches:
        drum_data[n.offset][p.midi] = 1
        print("note {} : {}: ".format(p.midi, n.offset))
```

```python
for n in o.parts[0].getElementsByClass(music21.note.Note):
    print("note {} : {}".format(n.pitch.midi, n.offset))
    if n.offset not in drum_data:
        drum_data[n.offset] = dict()
    drum_data[n.offset]['offset'] = n.offset
    drum_data[n.offset][n.pitch.midi] = 1
```

```python
'''
  drump_data 
    1.dataFrame ,
    2. NA 0  ,
    3. offset  
 offset note .
'''
import pandas as pd

drumDF = pd.DataFrame(list(drum_data.values()))
drumDF.fillna(0, inplace=True)
drumDF.sort_values('offset', inplace=True)

drumDF
```

```python
'''
RNN      (k) .
      .

1. 2 
2.   offset  [ 1/k, 2/k, ... , (k-1)/k]   k .
3.   k , RNN    (time window?) 1/k    .
    1. 0 1/k   drum        0 
    2.  ceil(max(offset)) .        .
'''

import math
from fractions import Fraction

#  k .
def getSmallestTempo(offset):
    offsetValues = offset[offset.notnull()].apply(lambda x: x - math.floor(x)).unique()
    offsetValues = pd.Series(offsetValues[0 < offsetValues])
    
    def validTempo(x, k):
        '''
        check: if x in [ 1/k , 2/k, ..., (k-1)/k ]
        '''
        return x in [Fraction(z, k) for z in range(1, k)]

    k = 1
    while True:
        k = k + 1
        isValidTempo = offsetValues.apply(validTempo , args=(k,))
        if isValidTempo.all() or k > 50:
            break
            
    print('found k = ', k)
    return k
k = getSmallestTempo(drumDF['offset'])
```

```python
'''
3.   k , RNN    (time window?) 1/k    .
    1. 0 1/k   drum        0 
    2.  ceil(max(offset)) .        .
'''

# 0  tick  
# tickList = [Fraction(z, k) for z in range(math.floor(drumDF['offset'].min() * k), math.ceil(drumDF['offset'].max() * k))]
tickList = [Fraction(z, k) for z in range(0, math.ceil(drumDF['offset'].max() * k))]

emptyTickList = list()
#  tick  
for tick in tickList:
    #  drumDF  tick   
    if len(drumDF.loc[drumDF['offset'] == tick]) == 0:
        #   tick  tick .
        emptyTickList.append(tick)

#   tick offset  DataFrame
df2 = pd.DataFrame(emptyTickList, columns=['offset'])

# drumpDF   tick  NA 0 .
drumTickDF = drumDF.append(df2).fillna(0)

#  offset  ,
# (0  min offset )
drumTickDF.sort_values('offset', inplace=True)
drumTickDF
```

```python
#    
for col in drumTickDF:
    if col != 'offset':
        drumTickDF[col] = drumTickDF[col].astype(int)
drumTickDF
```

```python
def getNotes(vec):
    '''
    get note from multi-label vector
    '''
    notes = []
    for idx in np.where(vec==1):
        if len(idx) > 0:
#             print(idx)
            for i in idx:
                notes.append(drumTickDF.columns[i + 1])
    return notes
```

```python
def getPrecision(predResult, y_labels):
    precision = 0.0
    matchCount = 0
    allCount = 0
    emptyCount = 0
    for j, (result, y_label) in enumerate(zip(predResult, y_labels)):
        predictVal = getNotes(np.round(result[-1]))
        yVal = getNotes(y_label[-1])
        if len(predictVal) > 0 and len(yVal) > 0:
            allCount = allCount + 1
            if len(predictVal) == len(yVal):
                matched = [i for i, k in zip(predictVal, yVal) if i == k]
                if len(matched) == len(yVal):
                    matchCount = matchCount + 1
        elif len(yVal) > 0 or len(predictVal) > 0:
            allCount = allCount + 1
        else:
    #         print('## result, y_label')
    #         print(result)
    #         print(y_label)
            emptyCount = emptyCount + 1

#         print(j, getNotes(np.round(result[-1])), getNotes(y_label[-1]))
    return (matchCount + emptyCount) / (allCount + emptyCount), matchCount / allCount, matchCount, allCount, emptyCount
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.contrib import rnn
import simplejson as json

# : https://github.com/hunkim/DeepLearningZeroToAll/blob/master/lab-12-4-rnn_long_char.py
# loss https://github.com/hunkim/DeepLearningZeroToAll/blob/master/lab-12-5-rnn_stock_prediction.py

data_dim = len(inputData[0])
hidden_size = len(inputData[0])
num_classes = len(inputData[0])
learning_rate = 0.1

print('sequence_length, iteration, loss, precision_non_empty, precision_empty, matchCount, allCount, emptyCount')
for seq_len in [50]:
# for seq_len in [50]:
    '''
    inputData X, Y  
        X: X[i] = inputData[i: i + sequence_length]  i X 
        Y: Y[i] = inputData[i + 1: i + 1 + sequence_length]  i Y 
    '''
    Xdata = np.array([inputData[i: i + seq_len] for i in range(len(inputData) - seq_len - 1)])
    Ydata = np.array([inputData[i + 1: i + 1 + seq_len] for i in range(1, len(inputData) - seq_len)])

    batch_size = len(Xdata)
    
    # X, Y   
    print(len(Xdata[0]), len(Xdata[-1]))
    print(len(Ydata[0]), len(Ydata[-1]))

    print(len(Xdata), len(Ydata))

    tf.reset_default_graph()
        
    X = tf.placeholder(tf.float32, [None, seq_len, data_dim])
    Y = tf.placeholder(tf.float32, [None, seq_len, data_dim])

    # Make a lstm cell with hidden_size (each unit output vector size)
    def lstm_cell():
        cell = rnn.BasicLSTMCell(hidden_size, state_is_tuple=True)
        return cell

    multi_cells = rnn.MultiRNNCell([lstm_cell() for _ in range(2)], state_is_tuple=True)
    print(multi_cells)
    # outputs: unfolding size x hidden size, state = hidden size
    outputs, _states = tf.nn.dynamic_rnn(multi_cells, X, dtype=tf.float32)
    print(outputs)
    # FC layer
    X_for_fc = tf.reshape(outputs, [-1, hidden_size])
    print(X_for_fc)
    outputs = tf.contrib.layers.fully_connected(X_for_fc, num_classes, activation_fn=None)
    print(outputs)
    
    # reshape out for sequence_loss
    outputs = tf.reshape(outputs, [batch_size, seq_len, num_classes])
    print(outputs)
    # outputs sigmoid    .
    #   0.5  Note 1       .
    pred = tf.nn.sigmoid(outputs)
    print(pred)

    logits_flat = tf.reshape(outputs, [-1, num_classes])
    print(logits_flat)
    targets = tf.reshape(Y, [-1, num_classes])
    print(targets)

    crossent = tf.nn.sigmoid_cross_entropy_with_logits(labels=targets, logits=logits_flat)
    print(crossent)
    crossent = tf.reduce_sum(crossent)
    print(crossent)
    # All weights are 1 (equal weights)
    weights = tf.ones([batch_size, seq_len])

    total_size = tf.reduce_sum(weights)
    total_size += 1e-12  # to avoid division by 0 for all-0 weights
    crossent /= total_size

    mean_loss = tf.reduce_mean(crossent)
    train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(mean_loss)

    sess = tf.Session()
    
    saver = tf.train.Saver()
    sess.run(tf.global_variables_initializer())
    for i in range(5):
        _, l, results, predResult = sess.run(
            [train_op, mean_loss, outputs, pred], feed_dict={X: Xdata, Y: Ydata})
        save_path = saver.save(sess, "./models/dr_rnn_model.ckpt")
        if i % 1 == 0:
            precision_empty, precision_non_empty, matchCount, allCount, emptyCount = getPrecision(predResult, Ydata)
            print(seq_len, i, l, precision_non_empty, precision_empty, matchCount, allCount, emptyCount)
    print(seq_len, i, l, precision_non_empty, precision_empty, matchCount, allCount, emptyCount)
```
